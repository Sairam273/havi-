```python
from tkinter import *
from tkinter import ttk
from tkcalendar import *
import mysql.connector

def error():
    error=Tk()
    error.geometry('100x100')
    error.configure(background="skyblue")
    Label(error,text="Invalid",bg="skyblue",fg="red").pack(fill=BOTH,expand=True)
def table():
    event=Tk()
    event.geometry("500x400")
    event.configure(background="skyblue")
    con=mysql.connector.connect(user="root",password="",host="localhost",database="havi")
    cur=con.cursor()
    cur.execute("select * from data")
    l=cur.fetchall()
    Label(event,text="Date",bg="skyblue",width="10",font=('Arial',16,'bold')).grid(row=0,column=0)
    Label(event,text="Event",bg="skyblue",width="10",font=('Arial',16,'bold')).grid(row=0,column=1)
    Label(event,text="Time",bg="skyblue",width="10",font=('Arial',16,'bold')).grid(row=0,column=2)
    for i in range(len(l)):
        for j in range(len(l[0])):
            e=Label(event,bg="skyblue",text=l[i][j],width="10")
            e.grid(row=i+1,column=j)
    Button(event,text="Back",bg="red",command=lambda:[event.destroy()]).grid(row=i+2,column=i+2)
    con.commit()
    cur.close()
    con.close()
        
def register(name,email,v,dob,mobile,password):
    con=mysql.connector.connect(user="root",password="",host="localhost",database="havi")
    cur=con.cursor()
    cur.execute("insert into users values(%s,%s,%s,%s,%s,%s)",(name.get(),email.get(),v.get(),dob.get(),mobile.get(),password.get()))
    con.commit()
    cur.close()
    con.close()
def data(date,event,h,m):
    con=mysql.connector.connect(user="root",password="",host="localhost",database="havi")
    cur=con.cursor()
    if (-1<h.get()<24) and  (-1<m.get()<60):
        con=mysql.connector.connect(user="root",password="",host="localhost",database="havi")
        cur=con.cursor()
        time=str(h.get())+':'+str(m.get())
        cur.execute("insert into data values (%s,%s,%s)",(date.get(),event.get(),time))
    else:
        error()
    con.commit()
    cur.close()
    con.close()
    
def listcreator():
    todo=Tk()
    todo.geometry('500x400')
    todo.configure(background="skyblue")
    event=StringVar()
    date=StringVar()
    h=IntVar()
    m=IntVar()
    Label(todo,text="TO-DO",bg="skyblue").place(x=200,y=0)
    Label(todo,text="Date",bg="skyblue").place(x=5,y=30)
    cal = DateEntry(todo,textvariable=date,date_pattern='yyyy-mm-dd',bg="skyblue").place(x=40,y=30)
    Label(todo,text="Hr",bg="skyblue").place(x=140,y=30)
    Spinbox(todo,width="5",textvariable=h,from_=0,to=23).place(x=160,y=30)
    Label(todo,text="Min",bg="skyblue").place(x=210,y=30)
    Spinbox(todo,width="5",textvariable=m,from_=0 ,to=59).place(x=240,y=30)
    Label(todo,bg='skyblue',text=" Enter To-Do ").place(x=5,y=70)
    Entry(todo,textvariable=event).place(x=90,y=70)
    Button(todo,bg="green",text = 'Submit',command =lambda:[data(date,event,h,m)]).place(x=220,y=70)
    Button(todo,bg="green",text = 'Table',command =lambda:[table()]).place(x=300,y=70)
    Button(todo,text="Back",bg="red",command=lambda:[todo.destroy(),main()]).place(x=350,y=70)
def signup():
    sign=Tk()
    sign.geometry('500x400')
    sign.configure(background="skyblue")
    name=StringVar()
    email=StringVar()
    dob=StringVar()
    mobile=IntVar()
    password=StringVar()
    Label(sign,text="Name",bg="skyblue").place(x=5,y=10)
    Entry(sign,textvariable=name).place(x=90,y=10)
    Label(sign,text="Email",bg="skyblue").place(x=5,y=40)
    Entry(sign,textvariable=email).place(x=90,y=40)
    Label(sign,text="Gender",bg="skyblue").place(x=5,y=70)
    v=StringVar()
    Radiobutton(sign,text="Male",variable=v,value="Male",bg="skyblue").place(x=90,y=70)
    Radiobutton(sign,text="Female",variable=v,value="Female",bg="skyblue").place(x=150,y=70)
    Label(sign,text="Date of Birth",bg="skyblue").place(x=5,y=100)
    cal = DateEntry(sign,textvariable=dob,date_pattern='dd-mm-yyyy',bg="skyblue").place(x=90,y=100)
    Label(sign,text="Mobile no.",bg="skyblue").place(x=5,y=130)
    Entry(sign,textvariable=mobile).place(x=90,y=130)
    Label(sign,text="Password",bg="skyblue").place(x=5,y=160)
    Entry(sign,textvariable=password,show="*").place(x=90,y=160)
    Button(sign,bg="green",text = 'Submit',command = lambda:[sign.destroy(),main(),register(name,email,v,dob,mobile,password)]).place(x=5,y=190)
    Button(sign,text="Back",bg="red",command=lambda:[sign.destroy(),main()]).place(x=90,y=190)
def validate(login,name,password):
    con=mysql.connector.connect(user="root",password="",host="localhost",database="havi")
    cur=con.cursor()
    cur.execute("select name,password from users")
    l=cur.fetchall()
    if (name.get(),password.get()) in l:
        login.destroy()
        listcreator()
    else:
        error()
    con.commit()
    cur.close()
    con.close()
    
def login():
    login=Tk()
    login.geometry('500x400')
    login.configure(background="skyblue")
    name=StringVar()
    password=StringVar()
    Label(login,text="LOGIN",bg="skyblue").place(x=200,y=0)
    Label(login,text="Username",bg="skyblue").place(x=5,y=30)
    Entry(login,textvariable=name).place(x=90,y=30)
    Label(login,text="Password",bg="skyblue").place(x=5,y=60)
    Entry(login,show="*",textvariable=password).place(x=90,y=60)
    Button(login,bg="green",text = 'Submit',command = lambda:[validate(login,name,password)]).place(x=5,y=90)
    Button(login,text="Back",bg="red",command=lambda:[login.destroy(),main()]).place(x=90,y=90)
def avalidate(admin,name,password):
    con=mysql.connector.connect(user="root",password="",host="localhost",database="havi")
    cur=con.cursor()
    if(name.get()==password.get()=='admin'):
        admin.destroy()
        details1=Tk()
        details1.geometry("500x400")
        details1.configure(background="skyblue")
        cur.execute("select * from users")
        l=cur.fetchall()
        Label(details1,text="Name",bg="skyblue",width="8",font=('Arial',16,'bold')).grid(row=0,column=0)
        Label(details1,text="Email",bg="skyblue",width="8",font=('Arial',16,'bold')).grid(row=0,column=1)
        Label(details1,text="gender",bg="skyblue",width="8",font=('Arial',16,'bold')).grid(row=0,column=2)
        Label(details1,text="dob",bg="skyblue",width="8",font=('Arial',16,'bold')).grid(row=0,column=3)
        Label(details1,text="mobile",bg="skyblue",width="8",font=('Arial',16,'bold')).grid(row=0,column=4)
        for i in range(len(l)):
            for j in range(len(l[0])-1):
                e=Label(details1,bg="skyblue",text=l[i][j],width="8")
                e.grid(row=i+1,column=j)
        Button(details1,text="Back",bg="red",command=lambda:[details1.destroy(),main()]).grid(row=i+2,column=0)
    else:
        error()
    con.commit()
    cur.close()
    con.close()
        
def admin():
    admin=Tk()
    admin.geometry('500x400')
    admin.configure(background="skyblue")
    name=StringVar()
    password=StringVar()
    Label(admin,text="ADMIN",bg="skyblue").place(x=200,y=0)
    Label(admin,text="Username",bg="skyblue").place(x=5,y=30)
    Entry(admin,textvariable=name).place(x=90,y=30)
    Label(admin,text="Password",bg="skyblue").place(x=5,y=60)
    Entry(admin,show="*",textvariable=password).place(x=90,y=60)
    Button(admin,bg="green",text = 'Submit',command = lambda:[avalidate(admin,name,password)]).place(x=5,y=90)
    Button(admin,text="Back",bg="red",command=lambda:[admin.destroy(),main()]).place(x=90,y=90)
def main():
    home=Tk()
    home.geometry('500x400')  
    home.configure(background="skyblue")
    Button(home,width="7",height="1",text = 'Sign Up',bg="skyblue",command =lambda:[home.destroy(),signup()]).place(x=200,y=170)
    Button(home,width="7",height="1", text = 'Login',bg="skyblue",command = lambda:[home.destroy(),login()]).place(x=200,y=200)
    Button(home,width="7",height="1", text = 'Admin',bg="skyblue",command = lambda:[home.destroy(),admin()]).place(x=200,y=230)
    home.mainloop()
main()
```
